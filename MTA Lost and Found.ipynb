```python

import pandas as pd

import requests

from bs4 import BeautifulSoup

import time
import datetime
```

```python
url = "http://advisory.mtanyct.info/LPUWebServices/CurrentLostProperty.aspx"

xml_data = requests.get(url).content

soup= BeautifulSoup(xml_data, "xml")
```

```python
claimed = soup.find("NumberOfItemsclaimed").text
```

```python
total = soup.find("NumberOfLostArticles").text
```

```python
categories = soup.find_all("Category")
item_list = []

for category in categories:
    group = category["Category"]
    
    subs = category.find_all("SubCategory")
    for sub in subs:
        item_dict = {}
        
        item = sub["SubCategory"]
        count = int(sub["count"])
        item_dict.update({
            "Category": group, 
            "SubCategory": item, 
            "amount": count
        })
        
        item_list.append(item_dict)
        
item_list.append({"Category" : "NumberOfItemsclaimed", "SubCategory" : "N/A", "amount" : claimed})
item_list.append({"Category" : "NumberOfLostArticles", "SubCategory" : "N/A", "amount" : total})
```

```python
df = pd.DataFrame(item_list)
```
